## Definition of the $1^{st}$ order ODE

The linear first-oder differential equation (linear in y and its derivative) can be write in the form 

$$
\frac{ dy }{ dx } =  - p(x) y + g(x) \quad i.e., \ y^{'} = f(x, y(x) )
$$
with the initial condition $ y(x_0) = y_0 $

Consider the following initial value problem $1^{st}$ order ode given by:
$$
t y^{'} + 2y = 4 t^2 \quad t > 0, \ y(1)=2
$$

Re-write it in the form $y^{'} = f(x, y(x))$
$$
y^{'} = - \frac{2}{t}y + 4t
$$

```python
from scipy.integrate import odeint 
import numpy as np
```

```python
def dy_dt(y, t):
    return -(2/t)*y + 4*t
```

```python
y0 = 2
t_start, t_end, t_step = 1, 2, 0.01
ts = np.arange(t_start, t_end, t_step)
ys = odeint(dy_dt, y0, ts)
```

```python
import matplotlib.pyplot as plt
```

```python
fig1 = plt.subplots( figsize=(10, 6) )
plt.plot(ts, ys)
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
```

## Definition of the $2^{nd}$ order ODE

The general linear second-order differential equation with independent variable $x$ and dependent variable $y$ is define as 
$$
\frac{d^2 y}{d x^2} = -p(x) y^{'} - q(x) y + g(x) \quad i.e., \ y^{''} = f( t, y, y^{'} )
$$
with the initial condition $y(x_0) = y_0 $ and $ y^{'}(x_0) = u_0 $

When $g(x)=0$, the equation is said to be **homogeneous** otherwise is **inhomogeneous**.

Solve the following IVP 
$$
y^{''} + y^{'} - 2y = x + sin(2x)
$$
where $ y(0) = 1, y^{'}(0) = 0 $.

Transforming the equation into system of DE. Let $x_1 = y, x_2 = y^{'}$ then
\begin{array}{l}
x^{'}_1 = x_2 \\
x^{'}_2 = 2 x_1 - x_2 + x + sin(2x)
\end{array}
Tranform matrix
$$
\begin{pmatrix}x^{'}_1  \\ x^{'}_2 \end{pmatrix} = 
\begin{pmatrix}0 & 1  \\ 2 &-1  \end{pmatrix} 
\begin{pmatrix} x_1  \\ x_2 \end{pmatrix} + 
\begin{pmatrix} 0  \\ x + sin(2x) \end{pmatrix}
$$
Then
$$
\frac{dP}{dx} = \begin{pmatrix}0 & 1  \\ 2 &-1  \end{pmatrix} P + \begin{pmatrix} 0  \\ x + sin(2x) \end{pmatrix}
$$
where $ P = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}, P(0) = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$.

```python
def dP_dx(P, x):
    return [ P[1], 2*P[0] - P[1] + x + np.sin(2*x) ]
```

```python
P0 = [1, 0]
t_start, t_end, t_steps= 0, 10, 0.01
ts = np.arange(t_start, t_end, t_steps)
```

```python
Ps = odeint(dP_dx, P0, ts)
ys = Ps[:,1]
```

```python
fig2 = plt.subplots( figsize=(12, 8) )
plt.plot(ts, ys)
plt.xlabel('x')
plt.ylabel('y(x)')
```
